## Task- 02

### Problem Statement
Explain the difference between Absolute and Relative positioning.



### Problem solution
**Relative Position**

In `position:relative` the element will be placed by default according to the normal flow of the document. when it gets `top`, `left`, `right`, and `bottom` properties to an element, it moves it's original location and it will not fill the gap left by the moved element.


how its works, let's take a look to the bellow code example-

```html
    <!--index.html-->
    <div class="relative-box">Box 1</div>
    <div id="b2" class="relative-box">Box 2</div>
    <div class="relative-box">Box 3</div>
```


defined three `div` element, every element given a shared class name `relative-box` and one element given unique id name `#b2`

```css
/*style.css*/
.relative-box {
        height: 50px;
        width: 50px;
        background-color: green;
        border: 1px solid red;
        position: absulute;
      }
      #b2 {
        top: 40px;
        left: 60px;
      }
```
All element given `position:relative`, only `Box 2` sets `top:40px` and `left:60px` properties and other two boxes are not sets the `top` and `left` property. In this case `Box 1` and `Box 3` will remain same of their original flow, and only `Box 2 ` will moved to top 40px and left 60px from the original location, beacause this element sets the top and left value

**Absolute Position**

The position absolute adjusts an element's position relative to the nearest ancestor/parent. if an element having no parent present, then the document body is used as parent. position absolute removes the element from normal document flow.

To understand the consepts let's break down some code example step by step-

```html
<!--index.htm-->
  <body>
    <div class="container">
      <p class="first-child">This box is absolute positioned</p>
      <p class="second-child">This box is not positioned</p>
    </div>
  </body>
  ```

  Inside the `body` element containig a child element `div` with class `container` and that `div` also containing two child element `p` with class name `first-child` and `second-child`

  ```css
  /*style.css*/
  .container {
        height: 300px;
        width: 300px;
        margin: 30px;
        background-color: red;
      }
      .first-child {
        width: 200px;
        height: 200px;
        background-color: aqua;
        position: absolute;
        top: 0;
        left: 0;
      }
      .second-child {
        width: 200px;
        height: 200px;
        background-color: blueviolet;
      }
```

class `.container` display a box with height and width 300px, margin: 30px and background color: red.
class `first-child` sets height and width 200px and background color aqua, this element gets an important property `position:absolute` and top and left 0, when this property applied to this element it will automatically positioned top-left corner of the browser viewport and it will determine `body` element as parent. In this case `first-child` has a parent `container` element, but `.container` is not explicitly positioned, it has default `position: static`. If no ancestor has a position other than static, it defaults to using the `body` (or the initial containing block), `second-child` element will take the place of absolute positioned element.

**Relative vs Absolute**


| Relative | Absolute |
|----------|----------|
| remains in the original flow | removed from the original document flow |
| create empty space left by moved element | not create empty space |
| positioned element from original positon | nearest positioned ancestor |

## Task 5:

### Problem statement
Explain the z-index, with a code example.

### Problem solution
In CSS the `z-index` property specifies stack order of an element. this property only works on elements with position value, if element has `static` postion then `z-index` will not work. the `z-index` can accept `integer` value, most of the time will see `integer` value like 1, 2, -99, an element with higher `integer` number will always priotize high and apear front of other lower number stacking order.

To understand the concepts let's see code examples-

```html
<!--index.html-->
    <div class="container">
      <div class="gray-box">Gray box</div>
      <div class="red-box">Red box</div>
      <div class="blue-box">Blue box</div>
    </div>
```
a `div` container contains three `div` elements.

```css
/*style.css*/
      .container {
        position: relative;
      }
      .container > div {
        width: 200px;
        height: 10rem;
        position: absolute;
      }
      .gray-box {
        background-color: gray;
        top: 0;
        left: 0;
        z-index: 1;
      }
      .red-box {
        background-color: red;
        top: 40px;
        left: 40px;
        z-index: 2;
      }
      .blue-box {
        top: 80px;
        left: 80px;
        background-color: blue;
        z-index: -1;
      }
```

In this css example, given every box to `z-index` property, `gray-box` gets value `1`, `red-box` 2 and `blue-box` -1. Here `red-box` will appear first because it gets higher integer value than others box. `gray-box` will appear after `red-box` and last `blue-box` appear behind red and gray box.
