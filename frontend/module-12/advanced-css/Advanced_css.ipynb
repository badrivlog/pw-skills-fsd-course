## Task-1

**Problem statement**

Explain the purpose of using the `var()` function in css. Also you have created two buttons with id named
primaryBtn and secondaryBtn which should be given background colors using the `var()` function. The color
code for primaryColor is `#00b7ff` and secondaryColor is `#6c757d`.

**Purpose of using the `var()` function in css**

According to the MDN docs- *The var() function in CSS is used to insert the value of a custom property (also called a CSS variable)*

for reusability purpose use css variable to declare the value with a specific name, for example (`--primaryColor: #00b7ff;`). Whenever need the primary color, just use css `var()` function to insert the value of custom property, for example,
```css
color: var(--primaryColor);
```
For mantainability, just change in one place (where it declared) without looking to all places where its used.

One great use cases of using `var()` function is theming. we can easily switch themes dark or light by changing a few custom properties.

````python
# index.html (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-1/index.html)
"""
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <link rel="stylesheet" href="style.css" />
    <title>Giving background color using CSS var() function</title>
  </head>
  <body>
    <button id="primaryBtn" class="btn">Primary button</button>
    <button id="secondaryBtn" class="btn">Secondary button</button>
  </body>
</html>
"""

# style.css (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-1/style.css)
"""
:root {
  --primaryColor: #00b7ff;
  --secondaryColor: #6c757d;
}

.btn {
  padding: 5px 10px;
  color: #fff;
  font-size: 18px;
  border: none;
}

#primaryBtn {
  background-color: var(--primaryColor);
}
#secondaryBtn {
  background-color: var(--secondaryColor);
}
"""
````

```css
:root {
  --primaryColor: #00b7ff;
  --secondaryColor: #6c757d;
}
```

In the above code example, declare two custom property value in the root of css file.

```css
#primaryBtn {
  background-color: var(--primaryColor);
}
#secondaryBtn {
  background-color: var(--secondaryColor);
}
```

given background color by using css var() function.

## Task- 2
**Problem statement**

Create a 3D cube using the transform property of CSS.

**Expected output**

<img src="https://drive.google.com/uc?export=view&id=12m7vhOcxTtDVrLtSQPs3OQnZDSzq8DET" width="200px"  alt="3d-cube"/>


````python
# index.html (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-2/index.html)
"""
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <link rel="stylesheet" href="style.css" />
    <title>Cube</title>
  </head>
  <body>
    <div class="container">
      <div class="cube">
        <div class="face top">Top</div>
        <div class="face bottom">Bottom</div>
        <div class="face left">Left</div>
        <div class="face right">Right</div>
        <div class="face front">Front</div>
        <div class="face back">Back</div>
      </div>
    </div>
  </body>
</html>
"""

# style.css (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-2/style.css)
"""
.container {
  width: 400px;
  height: 400px;
  perspective: 1000px;
  margin: 200px;
}
.cube {
  position: relative;
  width: 200px;
  height: 200px;
  transform-style: preserve-3d;
  transform: rotate3d(1, 2, 0, 30deg);
}
.face {
  width: 200px;
  height: 200px;
  background: #5279d6;
  border: 1px solid black;
  position: absolute;
  opacity: 0.5;
  display: flex;
  align-items: center;
  justify-content: center;
  font-size: 2rem;
}

.front {
  transform: translateZ(100px);
}
.back {
  transform: translateZ(-100px) rotateY(180deg);
}
.left {
  transform: translateX(-100px) rotateY(-90deg);
}
.right {
  transform: translateX(100px) rotateY(90deg);
}
.top {
  transform: translateY(-100px) rotateX(90deg);
}

.bottom {
  transform: translateY(100px) rotateX(-90deg);
}
"""
````

## Task-3
**Problem statement**

Create a simple circular loader which will rotate continuously to look like a loading screen on a website.

**Expected output**

<img src="https://drive.google.com/uc?export=view&id=1ILmWoDRoqbCzq-30U0hERQCg2FaMkUTa" width="200px"  alt="circular-loader"/>

````python
# index.html (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-3/index.html)
"""
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <link rel="stylesheet" href="style.css" />
    <title>Loader</title>
  </head>
  <body>
    <div id="loader"></div>
  </body>
</html>
"""

# style.css (https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-12/advanced-css/task-3/style.css)
"""
body {
  background-color: #eee;
  display: flex;
  align-items: center;
  justify-content: center;
  height: 100vh;
}

#loader {
  width: 50px;
  height: 50px;
  margin: 20px;
  border: 6px solid #fff;
  border-top: 6px solid #4444ee;
  border-radius: 50%;
  animation: spin 1.5s linear infinite;
}

@keyframes spin {
  0% {
    transform: rotate(0deg);
  }
  100% {
    transform: rotate(360deg);
  }
}
"""
````
