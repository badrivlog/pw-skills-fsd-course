Whenever we write some code we might need to store some values inside the memory (RAM). To store these values we can create buckets in the memory that hold these valuse. These buckets are called as `variables`. Variables can be created in the memory to store some data value.

### How can we create a variable in Javascript ?

To create a variable in JS, we have three dedicated keywords: `var`, `let` and `const`. Using these three keywords we can create a variable.

```javascript
var movieName = "Deadpool vs Wolverine";
let movieDescription = "The new wolverine saga"
const moviePrice = 350;
```

Here we can see we have created 3 values. The syntax works as follows:

* We first put the `var`, `let` or `const` keyword and put space after it. This indicates JS that we probaly want to create a variable.
* After the space we give the variable name.
* Then we put an `=` symbol denoting that we want to assign a value on the RHS of equala to the variable.
* The on the RHS of equals, we write the value we expect to get stored in the variable.
* Here `300` is a number and `Deadpool vs Wolverine` is a text. To create a text value, we just write the text and wrap it in a pair of double quotes.

### What is variable name ?

#### Updating values in a variable
If we have created variable using var or let we can update it's value by just reassigning something using equals symbol.
```JS
let x = 10
x = 11
```
Here value of `x` was 10 initially now it has been later updated to 11.

We cannot update a variable initialised by `const`. const  doesn't allow any reassignment or updates. If we try to do it and run the code, then it will throw us error.

### Rules to put variable names:

We cannot just give any name to our variable, there are rules that.
1. variable names must start with a letter or an underscore (_) or the dollar sign ($)
```JS
let $movieName = "Deadpool";
let _moviePrice = 500;
let movieDate = "4 Aug";
```
2. variable names cannot start with a number.
```JS
let 10data = 10;
```
The above code is invalid.
3. Variable names are case sensitive so that means if we have a variable with name `MovieName` or `moviename` or `MOVIENAME` etc all these variables will be considered different.
4. Variable names should have special characters apart from underscore or any number or dollar. That means we cannot use space, tab, hyphen etc in the variable names
```javascript
let x$y = 10;
```
This is a valid variable but
```javascript
let x-y = 99
```
This is an invalid variable.
5. Variable names cannot be a keyword. For example: we cannot make variable with name if else for etc as these are keywords.
6. Two different variables should not have same name.

## Comments in Javascript
Commets are extra piece of notes that we can write along with our code so that, if anyone is reading that code, they can understand what is going on in the code. Interestingly these notes will be completely avoided when JS will exucute the code. 

They are not going to serve any other purpose except providing annotations for the code.

#### How can we create a comment ?
In JS, we can create comments by following ways:
1. **Single line comment**: We can write `//` and then in that line we can write any random text, and this text will be considered as comment. With `//` we can only write comment in a single line, if we have to move to a new line while writing comments either in the line we again put `//` or we can use multiline comment.

2. **Multiline comment**: If we desire to write comments that go in multiple lines then instead of `//` we can use `/* */`. With this new symbol even if we go to a new line, then we don't have to manually put this as a comment and this new line will be automatically part of the comment.
```js
/*
this is a
multiline comment
with multiple lines
*/
```
So, here `/*` starts the multiline comment and `*/` ends it.

### BigInt
Sometimes we might have to use very big numbers, these big numbers are not by default supported by number datatype. The regular number datatype can handle values less than (2^53 - 1) and greater than -(2^53 - 1).


If we want to have a number beyond the range of `-(2^53-1)` -> (2^53-1) then we can use bigint.
We can create a bigint by just appending a `n` at the end of a very big number.

```js
let bigNumber = 7977397787857988477487n;
```

The big integers stored in a different way than our normal numbers. So any operation or calculations that we want to do on bigint is gonna be slightly different than normal number. For example: Addition substraction etc are all handled in different way.

Any boolean variable that we make can only get value true or false, as these two values are sufficient enough to represent two states.
