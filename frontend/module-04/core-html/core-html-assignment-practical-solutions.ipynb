## 4. Create an ordered list of HTML tags. Each list item must include the tag name and some information about the tag.

```python
# index.html (github file link- https://github.com/badrivlog/pw-skills-fsd-course/blob/main/frontend/module-04/core-html/qn-4/index.html)
"""
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>html tag list</title>
  </head>
  <body>
    <ol>
      <li>
        <strong>&lt;html&gt; :</strong>The root elememt that wraps all the
        content of html document.
      </li>
      <li>
        <strong>&lt;head&gt;:</strong> Contains metadata about the html
        document, like the title, links to stylesheets, and meta information.
      </li>
      <li>
        <strong>&lt;title&gt;:</strong> Specifies the title of the html
        document, which appears in the browser tab or title bar.
      </li>
      <li>
        <strong>&lt;body&gt;:</strong> Contains the visible content of the html
        document, such as text, images, and other elements.
      </li>
      <li>
        <strong>&lt;div&gt;:</strong> Creates division or section in html.
      </li>
      <li>
        <strong>&lt;h1&gt; to &lt;h6&gt;:</strong> For titles or subtitles.
      </li>
      <li><strong>&lt;p&gt;:</strong> Represents a paragraph of text.</li>
      <li>
        <strong>&lt;a&gt;:</strong> Defines a hyperlink, which is used to link
        one page to another or to specific parts of the same page.
      </li>
      <li><strong>&lt;img&gt;:</strong> Embeds image in the html document.</li>
      <li>
        <strong>&lt;span&gt;:</strong> An inline container element used for
        styling a small portion of text or other inline content.
      </li>
      <li><strong>&lt;ul&gt;:</strong> Defines unordered list.</li>
      <li><strong>&lt;ol&gt;:</strong> Defines ordered list.</li>
      <li>
        <strong>&lt;li&gt;:</strong> Defines a list item within either an
        unordered or ordered list.
      </li>
      <li>
        <strong>&lt;table&gt;:</strong> Represents a table in html document.
      </li>
      <li><strong>&lt;tr&gt;:</strong> Defines a table row.</li>
      <li><strong>&lt;td&gt;:</strong> Defines a cell in a table row.</li>
      <li><strong>&lt;th&gt;:</strong> Defines a header cell in a table.</li>
      <li>
        <strong>&lt;form&gt;:</strong> Defines an html form for collect user
        data.
      </li>
      <li>
        <strong>&lt;input&gt;:</strong> Represents an input field in a form,
        such as a text box, checkbox, or button.
      </li>
      <li><strong>&lt;button&gt;:</strong> Defines a clickable button.</li>
      <li>
        <strong>&lt;storng&gt;:</strong> Represents important text, rendered
        bold text.
      </li>
      <li>
        <strong>&lt;em&gt;:</strong> Represents emphasized text, rendered in
        italics.
      </li>
      <li>
        <strong>&lt;br&gt;:</strong> Creates a line break within a block of
        text.
      </li>
    </ol>
  </body>
</html>
"""
```
